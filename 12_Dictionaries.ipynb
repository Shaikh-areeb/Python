# 📘 Dictionaries in Python  

## 📌 Definition  

* A **dictionary** is a collection of **key-value pairs**  
* Each key is mapped to a value  
* Written using `{}`  

```python
d = {"name": "Areeb", "age": 22}
```

---

## 📏 Rules of Dictionary  

1. **Keys Must Be Unique**  

   ```python
   d = {"a": 1, "a": 2}  # ❌ last value overwrite
   ```

2. **Keys Must Be Immutable**  

   * Allowed: string, int, tuple  
   * Not allowed: list, set  

   ```python
   d = {[1,2]: "value"}  # ❌ Error
   ```

3. **Values Can Be Anything**  

   ```python
   d = {"a": [1,2,3], "b": "hello"}
   ```

4. **Mutable (Can Change)**  

   ```python
   d["age"] = 23
   ```

5. **Ordered (Python 3.7+)**  

   * Insertion order is maintained  

---

## ⚡ Basic Operations  

```python
d = {"name": "Areeb", "age": 22}

# Access
d["name"]
d.get("age")

# Add / Update
d["city"] = "Pune"

# Delete
del d["age"]
d.pop("name")

# Length
len(d)
```

---

## 💡 Tips & Tricks  

### 🔹 Important Methods  

```python
d.keys()
d.values()
d.items()
d.update({"age": 25})
d.clear()
```

---

### 🔹 Safe Access (Best Practice)  

```python
d.get("salary", 0)
```

---

### 🔹 Loop Through Dictionary  

```python
for key, value in d.items():
    print(key, value)
```

---

### 🔹 Dictionary Comprehension  

```python
squares = {x: x*x for x in range(5)}
```

---

### 🔹 Count Frequency (Very Important)  

```python
text = "hello"
freq = {}

for ch in text:
    freq[ch] = freq.get(ch, 0) + 1
```

---

### 🔹 Nested Dictionary  

```python
d = {
    "user": {
        "name": "Areeb",
        "age": 22
    }
}
```

---

## 🚀 When to Use Dictionary?  

* For **key → value mapping**  
* Fast lookup (O(1))  
* Counting / grouping data  
* API / JSON handling  

---

## ❌ Common Mistakes  

```python
d = {"a": 1}
d["b"]  # ❌ KeyError
```

```python
d = {[1,2]: "value"}  # ❌ unhashable key
```

---

* Dictionary = key-value pair collection

````python
subject_numbers = int(input('Enter the number of subjects'))

final_dict = {}

for i in range(subject_numbers):
    subject = input('Enter the Subject Name')
    marks = float(input('Enter The Marks'))
    final_dict[subject] = marks

print(final_dict)
````

````python
# Method 1

name = ['Areeb', 'Khizar', 'Faraz']
age = [23, 26, 31]

new_dict = dict(zip(name,age))

print(new_dict)
````

````python
# Method 2 

new_dict2 = {}

if len(name) == len(age):
    for i in range(len(name)):
        new_dict2[name[i]] = age[i]
else:
    print('len of list are not equal')

print(f'after merging two list -> {new_dict2}')
````

````python
print(final_dict)

total_marks = 0 

for i in final_dict.values():
    total_marks += i 

print(f'ttoal marks = {total_marks}')
````

````python
eg = 'Hello World'

freq = {}

for i in eg:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

freq
````

````python
student_dict = {
    'student_1' : [85, 45, 92, 78, 37],
    'student_2' : [85, 41, 92, 48, 47],
    'student_3' : [85, 65, 91, 88, 77],
    'student_4' : [85, 65, 92, 68, 51]
}

for key, value in student_dict.items():
    total = 0 
    for i in value:
        total += i
    percentage = (total / 500) * 100
    print(f'{key} has scored total marks of : {total} and secured : {percentage}%')
````

````python
user_subject = input('enter the subject')

if user_subject in final_dict:
    print(f'The Subject Exist in The Dictionary')
    print(final_dict[user_subject])
else:
    print('Subject not in dictionary')
````

````python
highest_marks = 0
highest_marks_student = ''

for name, marks in student_dict.items():
    total = sum(marks)
    if total > highest_marks:
        highest_marks = total
        highest_marks_student = name

print(f'The Highest marks secured by {highest_marks_student} with {highest_marks}')
````
