### **Escape Sequence**

| Escape Sequence | Definition |
|-----------------|------------|
| \n | Moves the cursor to a new line |
| \t | Inserts a horizontal tab space |
| \\ | Prints a backslash character |
| \' | Prints a single quote character |
| \" | Prints a double quote character |
| \r | Moves the cursor to the beginning of the line |
| \b | Removes the previous character (backspace) |
| \f | Inserts a form feed |
| \v | Inserts a vertical tab |

```python
print("Hello\nWorld")      # new line
print("Name:\tAreeb")      # tab space
print("This is \\ backslash")
print("He said \"Hi\"")
```

### **Variables**

Pyhton variables are simply containers for storing data values

#### Rules for Variables in Python

1. A variable name must start with a letter (a-z, A-Z) or an underscore (_).  
2. A variable name cannot start with a number.  
3. A variable name can only contain letters, numbers, and underscores (A-Z, a-z, 0-9, _).  
4. Variable names are case-sensitive (age and Age are different).  
5. Keywords (reserved words) cannot be used as variable names.  
6. No special characters like @, $, %, ! are allowed in variable names.  
7. Spaces are not allowed in variable names; use underscore instead.  
8. Variable names should be meaningful and descriptive.


| Variable Case | Example |
|---------------|---------|
| snake_case | student_name |
| camelCase | studentName |
| PascalCase | StudentName |
| UPPERCASE | MAX_LIMIT |

### **Input Function**
 Python input()  Rules & Tips (Quick Guide)
 Basic Rules

input() always returns a string.

Program execution pauses until user enters value + presses Enter.

You can show a message inside brackets  input("Enter name: ").

```python
# Type Conversion (Very Important)

age = int(input("Enter age: "))      # convert to integer
price = float(input("Enter price: ")) # convert to float

print(f'The user entered the age is {age}')
print(f'The user entered the price is {price}')
```

```python
# Storing the inputs

name = input("Enter your name: ")

print(f'your name is {name}')
```

```python
# Multiple Inputs

a, b = input("Enter two numbers: ").split()

print(f'the first value is {a} \nThe second Value is {b}')
```

 Tips to Remember

 Use strip() to remove spaces
name = input().strip()

 Validate before converting
if value.isdigit():

 Give clear prompts to user.

 Dont perform math directly on raw input() (string).

```python
# Common Mistake

x = input("5 + 5 = ")   # gives "10" as string
print(x + x)            # output: 1010 
```

```python
#  Correct Way

x = int(input("5 + 5 = "))
print(x + x)            # output: 20 
```

```python
# ask of number of games played in a tournament and calculate total ponts
# points ( 1 win = 4 points, 1 tie = 2 points)

games_played = int(input('Enter the number of games played'))
games_won = int(input('Enter the number of games won'))
games_loss = int(input('Enter the number of games Loss'))

tie_match = games_played - (games_won + games_loss)

print(f'The total match played = {games_played}')
print(f'Total Tie match are {tie_match}')
print(f'the final score is {(games_won * 4) + (tie_match * 2)}')
```
