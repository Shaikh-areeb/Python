# 🔁 For Loop – Summary, Rules & Tips

## 📌 Summary
- `for` loop is used to iterate over a sequence (list, tuple, string, range, etc.)
- It executes a block of code **for each element**
- No need to manually control index (unlike `while` loop)

---

## 🧠 Basic Syntax
```python
for variable in sequence:
    # code block
```

---

📏 Rules
- Always use `in` keyword → `for x in sequence`
- Colon `:` is **mandatory** after loop declaration
- Proper **indentation** is required (usually 4 spaces)
- Loop variable is **auto-updated** in each iteration
- Loop runs **till sequence ends**
- Sequence must be **iterable** (list, tuple, string, set, dict, range)
- Avoid modifying the sequence while iterating
- Loop variable remains available after loop (last value persists)

---

⚡ Important Keywords
- `break` → exits loop immediately  
- `continue` → skips current iteration  
- `pass` → placeholder (does nothing)  
💡 Tips
- Prefer `for` loop when iterations are known  
- Use `enumerate()` instead of manual indexing  
- Use list comprehension for cleaner code:  
  ```python
  squares = [x**2 for x in range(5)]

Keep loops simple and readable

Use conditions inside loop for filtering


---

## ❌ Common Mistakes
```markdown
- Missing `:` after `for`  
- Wrong indentation  
- Using non-iterable objects  
- Modifying list while looping

````python
start = int(input('Enter The Number'))
end = int(input('Enter The Number'))

print(f'start num : {start} and end num : {end}')

if start > end:
    print('Invalid')
elif (start == end):
    print('Invalid')
else:
    for num in range(start, end + 1):
        print(num, end = ' ') 
````

````python
num = []
count = 0

for i in range(1,101):
    if i % 7 == 0:
        num.append(i)
        count += 1

print(f'The total nums divisible by 7 from 1-100 are {count}')
print(num)
````

````python
num = []
count = 0

for i in range(1,201):
    if i % 7 == 0 and i % 6 == 0:
        num.append(i)
        count += 1

print(f'The total nums divisible by 6 and 7 from 1-200 are {count}')
print(num)
````

````python
# continue skips the iteration

for i in range(1,11):
    if i >= 5 and i <= 7:
        continue                # skips 5 - 7
    print(i, end= ' ')
````
