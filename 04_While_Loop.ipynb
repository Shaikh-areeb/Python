# 🔁 While Loop - Definition & Notes

## 📌 Definition
- A **while loop** is a control flow statement that repeatedly executes a block of code **as long as the condition is True**.
- The loop stops when the condition becomes False.

---

## 🔄 Syntax
```python
while condition:
    # code block
```

---

## 🧠 How It Works
- First, the condition is checked  
- If **True → code block executes**  
- Then the condition is checked again  
- This continues until the condition becomes False  

---

## ✅ Example
```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

### 🔹 Output
```
1
2
3
4
5
```

---

## ⚠️ Important Points
- The condition must eventually become False  
- Always update the variable inside the loop  
- Initialization + Condition + Update are necessary  

---

## 🔥 Infinite Loop Example (Avoid ❌)
```python
i = 1

while i <= 5:
    print(i)
```
👉 Here, `i` is never updated → loop will run forever  

---

## 🧩 While with Break
```python
i = 1

while i <= 10:
    if i == 5:
        break
    print(i)
    i += 1
```

---

## ⏭️ While with Continue
```python
i = 0

while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

---

## 💡 Tips & Best Practices
- ✔ Initialize variables before the loop  
- ✔ Clearly define the condition  
- ✔ Don’t forget increment/decrement  
- ✔ Avoid infinite loops  
- ✔ Use `break` for early exit  
- ✔ Use `while True` only when intentional (e.g., user input, APIs)

---

## 🚀 Real Use Cases
- Taking user input until valid input is received  
- API polling / data streaming  
- Retry logic until success  
- Game loops / real-time systems  

---

## 🆚 While vs For Loop
| While Loop | For Loop |
|-----------|---------|
| Condition-based | Iteration-based |
| Unknown iterations | Known iterations |
| Manual control | Automatic iteration |

---

## 🎯 Summary
- While loop runs **as long as the condition is True**  
- Gives full control over iteration  
- Powerful but must be used carefully to avoid infinite loops
