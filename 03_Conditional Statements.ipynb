# 🔹 Conditional Statements

## 📌 Definition
A **conditional statement** is a programming construct that allows execution of different code blocks based on whether a specified condition is **True or False**.

👉 In simple terms:  
It helps your program **make decisions**.

# 🧠 Types of Conditional Statements

## 🔸 1. if Statement
- Executes code only when the condition is true

## 🔸 2. if-else Statement
- Executes one block if condition is true
- Executes another block if condition is false

## 🔸 3. if-elif-else Statement
- Used when multiple conditions need to be checked
- Runs the first true condition block

## 🔸 4. Nested if Statement
- One `if` statement inside another `if`
- Used for complex decision-making

## 🔸 5. Ternary Operator (Short-hand)
- One-line conditional statement
- Used for simple conditions

---

# ⚡ Tips for Conditional Statements

- ✅ Keep conditions simple and easy to understand
- ✅ Avoid too many nested conditions
- ✅ Use `elif` instead of multiple `if` when possible
- ✅ Maintain proper indentation (important in Python)
- ✅ Use logical operators (`and`, `or`, `not`) wisely
- ✅ Write clean and readable conditions

---

# 🎯 Best Practices

- ✔️ Always handle edge cases
- ✔️ Use `else` for default conditions
- ✔️ Avoid redundant or repeated conditions
- ✔️ Test all possible scenarios
- ✔️ Keep code optimized and clean

```python
age = 5

if age > 18:
    print("You are an adult.")
```

```python
age = 18 

if age >= 18:
    print("You are an adult. \nYou can vote.")
else:
    print("You are a minor.")
```

```python
number_1 = int(input("Enter a number: "))
number_2 = int(input("Enter another number: "))

print(f"Number 1: {number_1}, Number 2: {number_2}")

if number_1 > number_2:
    print(f"{number_1} is greater than {number_2}.")
elif number_1 < number_2:
    print(f"{number_1} is less than {number_2}.")
else:
    print(f"{number_1} is equal to {number_2}.")
```

```python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(f"{number} is even.")
else:
    print(f"{number} is odd.")
```

```python
physics = int(input("Enter your Physics marks: "))
chemistry = int(input("Enter your Chemistry marks: "))

if physics >= 40 and chemistry >= 40:
    print("You have passed both subjects.")
else:
    print("You have failed one or both subjects.")
```

```python
marks = int(input('Enter the marks'))

print(f'marks : {marks}')

if marks > 100:
    print('Invalid')
elif marks >= 90:
    print(f'Grade A')
elif marks >= 80:
    print(f'Grade B')
elif marks >= 70:
    print(f'Grade C')
elif marks >= 60:
    print(f'Grade D')
else:
    print(f'Fail')
```

## 🔹 If-Else Binning Rules Tips

---

### ✅ 1. If-Else Binning Rule

- **If conditions are overlapping → use descending order**
  - Highest value / most restrictive condition first  
  - Example: `>=90`, `>=75`, `>=50`

- **Why?**
  - `if-elif` executes from top to bottom  
  - Once a condition is true, the rest are skipped

- **If ranges are non-overlapping → order is flexible**
  - Example: `50–75`, `75–90`, `90+`

---

### 💡 Golden Rule:
> **"Always write the most restrictive (highest) condition first"**
