# 🧩 Tuples in Python (Simple English)

## 📌 Definition

* A **tuple** is a collection of items.
* It is **ordered** and **cannot be changed** (immutable).
* Written using `()` brackets.

```python
t = (1, 2, 3)
```

---

## 📏 Rules of Tuples

1. **Ordered**

   * Items stay in the same order.
   * You can access them using index.

   ```python
   t[0]  # 1
   ```

2. **Immutable (Cannot Change)**

   * You cannot update or delete items.

   ```python
   t[0] = 10  # ❌ Error
   ```

3. **Duplicates Allowed**

   ```python
   t = (1, 1, 2)
   ```

4. **Different Data Types Allowed**

   ```python
   t = (1, "Areeb", 3.5)
   ```

5. **Single Item Tuple**

   * Must use comma

   ```python
   t = (5,)  # correct
   ```

6. **Nested Tuple Allowed**

   ```python
   t = (1, (2, 3))
   ```

---

## ⚡ Basic Operations

```python
t = (10, 20, 30)

# Access
t[1]        # 20

# Slicing
t[1:]       # (20, 30)

# Length
len(t)

# Add tuples
t + (40,)

# Repeat
t * 2
```

---

## 💡 Tips & Tricks

### 🔹 Packing & Unpacking

```python
t = 1, 2, 3   # packing

a, b, c = t   # unpacking
```

### 🔹 Swap Values Easily

```python
a, b = 5, 10
a, b = b, a
```

### 🔹 Return Multiple Values

```python
def fun():
    return 10, 20

x, y = fun()
```

### 🔹 Faster than List

* Tuples are usually faster and use less memory.

### 🔹 Use as Dictionary Key

```python
d = {(1, 2): "value"}
```

---

## 🚀 When to Use Tuple?

* When data **should not change**
* When you want **fixed values**
* When returning **multiple values from function**

---

## ❌ Common Mistakes

```python
t = (5)   # ❌ not a tuple
t = (5,)  # ✅ tuple
```

---

## 🧠 Summary

* Tuple = ordered + cannot change
* Good for fixed data
* Fast and safe
